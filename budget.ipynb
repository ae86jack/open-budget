```python
def parse_float(value):
    if not value:
        return None
    try:
        print(type(value), str(value))
        if np.isnan(value):
            return None
        v_f = float(value.replace(',', ''))
        return v_f
    except TypeError:
        return None
    except ValueError:
        return None
```
